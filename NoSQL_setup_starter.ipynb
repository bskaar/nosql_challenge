2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
query = {
    "BusinessType": "Restaurant/Cafe/Canteen"
}

projection = {
    "_id": 0,
    "BusinessTypeID": 1,
    "BusinessType": 1
}

result = establishments.find(query, projection)
for document in result:
    pprint(document)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen"
query = {
    "BusinessType": "Restaurant/Cafe/Canteen"
}

projection = {
    "_id": 0,
    "BusinessTypeID": 1
}

business_type = establishments.find_one(query, projection)
if business_type:
    business_type_id = business_type["BusinessTypeID"]

    # Update the new restaurant with the BusinessTypeID
    update_query = {
        "BusinessName": "Penang Flavours"
    }

    update_data = {
        "$set": {
            "BusinessTypeID": business_type_id
        }
    }

    establishments.update_one(update_query, update_data)
    print("Restaurant updated successfully!")
else:
    print("BusinessTypeID not found for 'Restaurant/Cafe/Canteen'. Update failed.")
```

```python
# Confirm that the new restaurant was updated
updated_restaurant = establishments.find_one({"BusinessName": "Penang Flavours"})
pprint(updated_restaurant)
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
count_documents = establishments.count_documents({"LocalAuthorityName": "Dover"})
print(f"Number of documents with LocalAuthorityName as 'Dover': {count_documents}")
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
update_query = {
    "$or": [
        {"longitude": {"$type": "string"}},
        {"latitude": {"$type": "string"}}
    ]
}
```

```python
# Update and use $toDouble to convert the string values to double/decimal type.
update_data = {
    "$set": {
        "longitude": {"$toDouble": "$longitude"},
        "latitude": {"$toDouble": "$latitude"}
    }
}
```

```python
# Use update_many to convert latitude and longitude to decimal numbers.
update_result = establishments.update_many(update_query, update_data)
print(f"Number of documents updated: {update_result.modified_count}")
```

```python
# Verify the changes by retrieving and displaying one of the updated documents.
updated_document = establishments.find_one({"longitude": {"$type": "double"}})
pprint(updated_document)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
update_query = {"RatingValue": {"$type": "string"}}
update_data = {"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}
update_result = establishments.update_many(update_query, update_data)
print(f"Number of documents updated: {update_result.modified_count}")
```

```python
# Check that the coordinates and rating value are now numbers
updated_rating = establishments.find_one({"RatingValue": {"$type": "int"}})
pprint(updated_rating)
```
